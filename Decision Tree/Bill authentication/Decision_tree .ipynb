```python
#Importing required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
#Importing the dataset
df = pd.read_csv("bill_authentication.csv")
print(df.head())
```

```python
#Info of the Data
print(df.info())
```

```python
# Column Names
print(df.columns)
```

```python
#Data Description 
print(df.describe())
```

```python
#Histogram Representation
df.hist(bins=50, figsize=(20,15),xlabelsize=6,ylabelsize=6)
plt.show()
```

```python
#Heat Map Representation
d = df.copy()
plt.figure(figsize=(23,20))
sns.heatmap(data = d.corr(),annot = True,cmap="PuOr",fmt=".2f")
plt.show()
```

```python
#Countplot of the class
sns.countplot(x='Class',data=df)
plt.show()
```

```python
x = df.drop('Class',axis=1)
y = df['Class']
print(x.head())
print(y.head())
```

```python
#Trainning the Data 
dt = DecisionTreeClassifier()
x1,x2,y1,y2 = train_test_split(x,y,test_size=0.2,random_state=0)
dt.fit(x1,y1)
pred = dt.predict(x2)
```

```python
#plotting the Decision Tree
plot_tree(dt)
```

```python
#Accuracy Prediction 
acc = accuracy_score(y2,pred)
print(f"The Accuracy is {acc}")
```

```python
#Confusion matrix and Classification Report
print(confusion_matrix(y2,pred))
print(classification_report(y2,pred))
```
